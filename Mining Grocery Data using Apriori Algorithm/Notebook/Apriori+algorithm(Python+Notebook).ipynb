```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('Grocery.csv', header=None)
df.head()
```

```python
df.shape
```

```python
basket_sets = pd.get_dummies(df)
```

```python
basket_sets.head(5)
```

```python
basket_sets.tail(5)
```

```python
#applying apriori algorithm to generate rules
frequent_itemsets = apriori(basket_sets, min_support=0.003, use_colnames=True)
frequent_itemsets.tail
frequent_itemsets.shape
```

```python
#calculating confidence and lift for the genrated rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
#selecting the rules with lift >1
rules[ (rules['lift'] >= 1 ) &
       (rules['confidence'] >= 0.5) ]
```
